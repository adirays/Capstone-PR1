```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Data loading
df = pd.read_csv("C:/Users/91810/Downloads/manufacturing_dataset_1000_samples.csv")
```

```python
#Exploratory Data Visualization
plt.figure(figsize=(6,4))
sns.histplot(df['Parts_Per_Hour'], kde=True)#parts_per_hour 
plt.title("Distribution of Parts Per Hour")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Efficiency_Score'], kde=True)#efficiency_score histogram
plt.title("Distribution of Efficiency Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Parts_Per_Hour', y='Efficiency_Score', data=df)
plt.title("Parts Per Hour vs Efficiency Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Cycle_Time', y='Efficiency_Score', data=df)
plt.title("Cycle Time vs Efficiency Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Machine_Utilization', y='Parts_Per_Hour', data=df)
plt.title("Machine Utilization vs Parts Per Hour")
plt.show()
```
